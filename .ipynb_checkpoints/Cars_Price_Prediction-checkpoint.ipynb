```python
def identify_outliers(group):
    q1 = group['price'].quantile(0.25)
    q3 = group['price'].quantile(0.75)
    iqr = q3 - q1
    lb = q1 - 1.5 * iqr
    ub = q3 + 1.5 * iqr

    # Filter rows where price is outside the bounds
    outliers = group[(group['price'] < lb) | (group['price'] > ub)]
    return outliers
```

```python
outliers_df = df.groupby('make', group_keys=False).apply(identify_outliers)
```

```python
outliers_df
```

```python
def replace_outliers(group):
    q1 = group['price'].quantile(0.25)
    q3 = group['price'].quantile(0.75)
    iqr = q3 - q1
    lb = q1 - 1.5 * iqr
    ub = q3 + 1.5 * iqr
    median = group['price'].median()

    # Replace outliers with median
    group['price'] = group['price'].apply(lambda x: median if x < lb or x > ub else x)
    return group

# Apply by 'make'
df = df.groupby('make', group_keys=False).apply(replace_outliers)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
df[numerical_cols] = imputer.fit_transform(df[numerical_cols])

# Plot KDE for each numerical column
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.kdeplot(data=df, x=col, fill=True, color="blue")
    plt.title(f"KDE Plot of {col}")
    plt.show()
```
