```python
#We're now applying the IQR method to remove outliers. First,
#we identify where the outliers are locatedin this case, within the car dataset, 
#the outliers correspond to entries with unusually high charges.
plt.figure(figsize=(12,6))
sns.boxplot(data=df,x='price',y='make')
plt.title('distribution of make and price')
plt.show()
```

```python
#IQR methods
def identifying_outliers(group):
    q1=group['price'].quantile(0.25)
    q3=group['price'].quantile(0.75)
    IQR=q3-q1
    lb=q1-1.5*IQR
    ub=q3+1.5*IQR
    outliers=group[(group['price']<lb)|(group['price']>ub)] 
    return outliers[['make','price']]
```

```python
outliers=df.groupby('make').apply(identifying_outliers)
print(outliers)
```

```python
def remove_outliers(group):
    q1 = group['price'].quantile(0.25)
    q3 = group['price'].quantile(0.75)
    IQR = q3 - q1
    lb = q1 - 1.5 * IQR
    ub = q3 + 1.5 * IQR
    median = group['price'].median()
    # Replace outliers with median
    group['price'] = group['price'].apply(lambda x: median if x < lb or x > ub else x)
    
    return group
```

```python
df=df.groupby('make').apply(remove_outliers)
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df,x='price',y='make')
plt.title('distribution of make and price')
plt.show()
```

```python
x=df.drop(columns=['price'],axis=1)
y=df['price']
```

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x_scaled =st.fit_transform(x.iloc[::-1])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape,y_train.shape)
print(y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model=lr.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
mse=mean_squared_error(y_test,ypred)
rmse=np.sqrt(mse)
mae=mean_absolute_error(y_test,ypred)
r2=r2_score(y_test,ypred)
print(f"MSE:  {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAE:  {mae:.2f}")
print(f"R:   {r2:.4f}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
rf = RandomForestRegressor(random_state=42)

# Hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300],   # number of trees
    'max_depth': [None, 10, 20, 30],  # depth of each tree
    'min_samples_split': [2, 5, 10],  # min samples required to split a node
    'min_samples_leaf': [1, 2, 4]     # min samples per leaf node
}

# Grid Search with 3-fold CV
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3,
    scoring='r2',
    n_jobs=-1,
    verbose=2
)

grid_search.fit(x_train, y_train)

# Best parameters
print("Best Parameters:", grid_search.best_params_)

# Best model
best_rf = grid_search.best_estimator_

# Evaluate on test set
y_pred = best_rf.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"Best RMSE: {rmse:.2f}")
print(f"Best R:   {r2:.4f}")
```

```python
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import mean_squared_error, r2_score

# Lasso
lasso = Lasso(alpha=0.1)
lasso.fit(x_train, y_train)
y_pred_lasso = lasso.predict(x_test)

# Ridge
ridge = Ridge(alpha=1.0)
ridge.fit(x_train, y_train)
y_pred_ridge = ridge.predict(x_test)

# Evaluate
for name, y_pred in [("Lasso", y_pred_lasso), ("Ridge", y_pred_ridge)]:
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    print(f"{name} RMSE: {rmse:.2f}, R: {r2:.4f}")
```
